### Step 2. Download the dataset to your computer and unzip it.

```python
import kagglehub
import pandas as pd
import os

dataset_path = kagglehub.dataset_download("openfoodfacts/world-food-facts")

file_path = os.path.join(dataset_path, "en.openfoodfacts.org.products.tsv")

df = pd.read_csv(file_path, sep="\t", low_memory=False)

print(df.head())
```
