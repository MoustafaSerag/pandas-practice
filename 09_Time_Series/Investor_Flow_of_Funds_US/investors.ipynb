### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv).

```python
df=pd.read_csv('https://raw.githubusercontent.com/datasets/investor-flow-of-funds-us/master/data/weekly.csv')
```

### Step 3. Assign it to a variable called

```python
investors=df
investors.head()
```

### Step 4.  What is the frequency of the dataset?

```python
investors.head()
```

```python
investors.dtypes
```

```python
investors['Date'] = pd.to_datetime(investors['Date'])
```

```python
investors = investors.set_index('Date')
```

```python
investors = investors.sort_index()
```

```python
pd.infer_freq(investors.index)
```

```python
investors.resample('W').sum()
```

```python
investors.resample('M').mean()
```
